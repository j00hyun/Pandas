# Handling Missing Values

## 데이터 미리보기

```python
import pandas as pd
import numpy as np

# 실습 데이터: 미국프로풋볼(NFL) 경기
nfl_data = pd.read_csv("Data/NFL Play by Play 2009-2017 (v4).csv")

np.random.seed(0)
```

```python
# 데이터 미리보기
# 결측값이 많은 것을 볼 수 있음

nfl_data.head()
```

## 결측값 개수 확인

```python
# 각 column 별로 결측값이 몇 개 있는지 확인
missing_values_count = nfl_data.isnull().sum()

# 처음 10개 column의 결측값 개수 확인
missing_values_count[0:10]
```

```python
# 전체 데이터 중 얼마나 많은 비율이 비어 있는지 계산

# 얼마나 많은 결측값을 가지고 있는지 계산
total_cells = np.product(nfl_data.shape)
total_missing = missing_values_count.sum()

# 결측값의 비율을 퍼센트로 반환
percent_missing = (total_missing / total_cells) * 100
print(percent_missing)
```

## 결측값이 생긴 이유
1. **단순히 기록되지 않은 경우**
   - TimeSecs (플레이 시점의 경기 잔여 시간)은 기록이 누락된 경우로 보임
   - 이런 경우에는 추정을 통해 값을 채우는 게 가능
2. **애초에 존재하지 않는 경우**
   - PenalizedTeam (반칙을 한 팀)은 반칙이 없으면 존재하지 않는 값
   - 이럴 때는 그냥 NaN으로 두거나 "None", "Neither" 같은 값을 대체

## 결측값 제거히기

```python
# 결측값이 있는 row를 통째로 삭제
# 여기선 모든 row가 결측값을 포함하므로 0개의 행만 남음

nfl_data.dropna()
```

```python
# 결측값이 있는 column만 제거

columns_with_na_dropped = nfl_data.dropna(axis=1)
columns_with_na_dropped.head()
```

## 결측값 채우기

```python
# 먼저 보기 좋게 작은 부분만 가져오기

subset_nfl_data = nfl_data.loc[:, 'EPA':'Season'].head()
subset_nfl_data
```

```python
# NaN을 모두 0으로 채움

subset_nfl_data.fillna(0)
```

```python
# 후속값으로 채우기
# bfill: 뒤(row + 1)의 값으로 결측치 채움
# fillna(0): 그래도 남은 결측치는 0으로 채움

subset_nfl_data.fillna(method='bfill', axis=0).fillna(0)
```
