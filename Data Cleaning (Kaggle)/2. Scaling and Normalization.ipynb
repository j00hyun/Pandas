```python
import pandas as pd
import numpy as np

# Box-Cox Transformation
from scipy import stats

# Min-Max Scaling
from mlxtend.preprocessing import minmax_scaling

# 시각화
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(0)
```

## Scaling vs Normalization
- Scaling: 데이터의 범위를 바꿈 (데이터의 모양은 그대로)
- Normalization: 데이터의 모양을 바꿈 (정규분포 형태로 변환)

## Scaling

```python
# 데이터의 범위가 기존에는 0-8 정도였다면, 스케일링 후에는 0-1 사이로 변함

# 1,000개의 데이터를 지수분포(Exponential)로 생성
original_data = np.random.exponential(size=1000)

# Min-Max 스케일링 (0~1 범위로)
scaled_data = minmax_scaling(original_data, columns=[0])

# 비교를 위해 시각화
fig, ax = plt.subplots(1, 2, figsize=(15, 3))
sns.histplot(original_data, ax=ax[0], kde=True)
ax[0].set_title("Original Data")
sns.histplot(scaled_data, ax=ax[1], kde=True)
ax[1].set_title("Scaled Data")
plt.show()
```

## Normalization

```python
# 변환 전: L자형 (비대칭, 한쪽으로 치우친 분포)
# 변환 후: 종 모양 곡선으로 변환 (대칭적인 정규분포 형태)

# Box-Cox 변환
normalized_data = stats.boxcox(original_data)

# 시각화 비교
fig, ax = plt.subplots(1, 2, figsize=(15, 3))
sns.histplot(original_data, ax=ax[0], kde=True)
ax[0].set_title("Original Data")
sns.histplot(normalized_data[0], ax=ax[1], kde=True)
ax[1].set_title("Normalized Data")
plt.show()
```
