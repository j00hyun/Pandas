## Maps
기존 데이터를 새 형식으로 변환하거나 새 값을 만드는 작업에 사용

### map()
- Series의 각 원소에 함수를 적용
- 원본 데이터를 바꾸지 않고 새 Series를 반환

```python
# 와인 점수를 평균 기준으로 중심화 (평균을 0으로 맞춤)

review_points_mean = reviews.points.mean()
reviews.points.map(lambda p: p - review_points_mean)
```

### apply()
- DataFrame 전체에 함수를 적용 (각 row 또는 column 단위로 적용 가능)
- 원본 데이터를 유지한 채 새 DataFrame 반환

```python
# 모든 row(axis='columns')의 points 값을 평균 기준으로 중심화 후 새 DataFrame 반환

def remean_points(row):
    row.points = row.points - review_points_mean
    return row

reviews.apply(remean_points, axis='columns')
```

```python
# 기존 데이터는 변하지 않음

reviews.head(1)
```

### Pandas의 연산자 기반 변환

```python
# Series 전체에 대해 한번에 연산 가능

review_points_mean = reviews.points.mean()
reviews.points - review_points_mean
```

```python
# Series끼리도 같은 길이라면 직접 연산 가능
# country 와 region_1 결합

reviews.country + " - " + reviews.region_1
```
