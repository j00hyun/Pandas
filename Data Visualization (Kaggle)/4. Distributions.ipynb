# Distributions

## 환경 설정

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

## 데이터 불러오기
### Iris 붓꽃 3종류 (각 50개씩, 총 150개 샘플)에 대한 측정값
- **row** : 한 송이 꽃
- **Sepal Length (cm)** : 꽃받침의 길이
- **Sepal Width (cm)** : 꽃받침의 폭
- **Petal Length (cm)** : 꽃잎의 길이
- **Petal Width (cm)** : 꽃잎의 폭
- **Species** : 종 이름 (Iris-setosa, Iris-versicolor, Iris-virginica)

```python
iris_filepath = "Data/iris.csv"
iris_data = pd.read_csv(iris_filepath, index_col="Id")
iris_data.head()
```

## 히스토그램 (Histogram)
- 데이터를 구간별로 나눠서 빈도 수를 박대로 표현하는 그래프

```python
# X축: 꽃잎 길이 (Petal Length)
# Y축: 해당 길이를 가진 꽃의 개수 (count)
# 막대 높이 = 빈도 (frequency)
sns.histplot(iris_data['Petal Length (cm)'])

plt.show()
```

## 밀도 그래프 (Density plots)
- 히스토그램을 부드럽게 연결한 버전 -> 분포의 형태를 부드럽게 파악 가능

```python
# fill=True : 곡선 아래 부분 색깔로 채우기
sns.kdeplot(data=iris_data['Petal Length (cm)'], fill=True)

plt.show()
```

## 이변량 밀도 그래프 (2D KDE Plot)
- 두 개의 연속형 변수 간의 분포 관계 표현

```python
# 두 변수 간의 상관관계를 시각화

# X축: 꽃잎 길이, Y축: 꽃받침 폭
# 위쪽, 오른쪽에 각각 1D KDE 곡선이 자동 표시됨
sns.jointplot(
    x=iris_data['Petal Length (cm)'],
    y=iris_data['Sepal Width (cm)'],
    kind="kde"
)

plt.show()
```

## 색상 구분 그래프 (Color-coded plots)
- 각 꽃 종에 따라 다른 색상으로 분포를 비교

```python
# 히스토그램
    # Iris-setosa: 꽃잎이 짧은 쪽 (1~2cm)
    # Iris-versicolor: 중간 길이 (3~5cm)
    # Iris-virginica: 긴 쪽 (5~7cm)

# x: X축 변수 이름
# hue: 색상으로 구분할 범주형 변수
sns.histplot(data=iris_data, x='Petal Length (cm)', hue='Species')

plt.title("Histogram of Petal Lengths, by Species")
plt.show()
```

```python
# 커널 밀도 그래프 (KDE)
    # 꽃잎 길이가 2cm 미만이면 거의 확실히 Iris-setosa

sns.kdeplot(
    data=iris_data,
    x='Petal Length (cm)',
    hue='Species',
    fill=True
)
plt.title("Distribution of Petal Lengths, by Species")
plt.show()
```
