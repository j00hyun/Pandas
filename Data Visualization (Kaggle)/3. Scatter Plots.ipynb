# Scatter Plots

## 환경 설정

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

## 데이터 불러오기
- **age** : 나이
- **sex** : 성별
- **bmi** : 체질량지수 (Body Mass Index)
- **children** : 자녀 수
- **smoker** : 흡연 여부 ("yes", "no")
- **region** : 거주 지역
- **charges** : 보험료 (연간 청구 금액)

```python
insurance_filepath = "Data/insurance.csv"
insurance_data = pd.read_csv(insurance_filepath)
insurance_data.head()
```

## 산점도 (Scatter Plot)
- BMI와 보험료의 관계 시각화

```python
# x축: BMI, y축: 보험료 (charges)
# BMI가 높을수록 보험료가 높아지는 양의 상관관계 (positive correlation)
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'])

plt.show()
```

```python
# 회귀선이 포함된 산점도 (Regression Plot)
# 데이터의 전반적인 경향(trend)을 보여줌

sns.regplot(x=insurance_data['bmi'], y=insurance_data['charges'])
plt.show()
```

## 색상 구분 산점도 (Color-coded Scatter Plot)
- 흡연자와 비흡연자를 색으로 구분

```python
# 흡연이 보험료 상승의 핵심 요인 중 하나임을 보여줌
    # 비흡연자는 BMI가 높아질수록 보험료가 조금 증가
    # 흡연자는 BMI가 조금만 높아도 보험료가 급격히 상승

# hue='smoker' : 색상으로 그룹을 구분 ("yes" vs "no")
sns.scatterplot(x="bmi", y="charges", hue="smoker", data=insurance_data)

plt.show()
```

```python
# 흡연자 vs 비흡연자의 회귀선 각각 표시
    # 흡연자의 회귀선(기울기)이 훨씬 가파름 : BMI가 높을수록 보험료 급상승
    # 비흡연자는 완만한 증가 : 건강 위험이 덜함

sns.lmplot(x="bmi", y="charges", hue="smoker", data=insurance_data)
plt.show()
```

# 범주형 산점도 (Categorical Scatter Plot)
- x축이 범주형 데이터(“smoker”)일 때 산점도

```python
# x축: 흡연 여부, y축: 보험료
# 흡연자는 전체적으로 보험료가 훨씬 높음
# 비흡연자는 낮은 보험료 구간에 밀집
sns.swarmplot(x=insurance_data['smoker'], y=insurance_data['charges'])

plt.show()
```
