# Bar Charts and Heatmaps

## 환경 설정

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

## 데이터 불러오기
- 2015년 항공사 지연 통계
- **Month** : 월 (1월~12월)
- **AA, AS, B6, DL, ...** : 항공사 코드
- **값** : 각 항공사의 월별 평균 도착 지연 시간 (분 단위) -> 음수: 일찍 도착, 양수: 지연 도착

```python
flight_filepath = "Data/flight_delays.csv"

# index_col="Month" : “Month” column을 index로 지정
flight_data = pd.read_csv(flight_filepath, index_col="Month")
```

## 데이터 확인

```python
flight_data
```

## 막대그래프 (Bar chart)
- Spirit Airlines(NK)의 월별 평균 도착 지연 시간을 시각화
- 9~11월에는 비교적 항공편이 시간 맞춰 잘 운항됨

```python
# 그래프 크기 설정 (가로 10, 세로 6 인치)
plt.figure(figsize=(10,6))

# 그래프 제목 추가
plt.title("Average Arrival Delay for Spirit Airlines Flights, by Month")

# Seaborn으로 막대그래프 생성
# X축 → 월(Month)
# Y축 → Spirit Airlines의 평균 지연 시간
sns.barplot(x=flight_data.index, y=flight_data['NK'])

# Y축 이름 지정
plt.ylabel("Arrival delay (in minutes)")

plt.show()
```

## 히트맵 (Heatmap)
- 모든 항공사의 월별 평균 도착 지연 시간 한눈에 보기
- 각 셀 색깔은 평균 지연 시간에 따라 다르게 칠해짐
- **밝은 색** : 지연이 많음 (양수)
- **어두운 색** : 일찍 도착 (음수)
- 9~11월의 셀들이 전체적으로 어두움 → 대부분 항공사들이 가을에는 시간 준수

```python
plt.figure(figsize=(14,7))
plt.title("Average Arrival Delay for Each Airline, by Month")

# 히트맵(heatmap) 생성
# data=flight_data : 전체 데이터를 시각화
# annot=True : 각 셀에 실제 숫자 표시
sns.heatmap(data=flight_data, annot=True)

# X축 이름 추가
plt.xlabel("Airline")

plt.show()
```
