# Line Charts

## 환경 설정

```python
import pandas as pd

# pandas의 날짜 데이터를 matplotlib이 인식할 수 있도록 연결
pd.plotting.register_matplotlib_converters()

import matplotlib.pyplot as plt

# 그래프를 주피터 노트북 안에 바로 표시하기
%matplotlib inline

import seaborn as sns
```

## 데이터 불러오기
- Spotify의 글로벌 일일 스트리밍 수
- 발매 전은 NaN으로 표시됨

```python
spotify_filepath = "Data/spotify.csv"

# index_col="Date" : 날짜를 인덱스로 지정
# parse_dates=True : 문자열 날짜를 datetime 타입으로 자동 변환
spotify_data = pd.read_csv(spotify_filepath, index_col="Date", parse_dates=True)
```

## 데이터 확인

```python
# 쳣 5 row

spotify_data.head()
```

```python
# 마지막 5 row

spotify_data.tail()
```

## 데이터 시각화

```python
# 라인 차트

sns.lineplot(data=spotify_data)
plt.show()
```

```python
# 그래프 크기와 제목 추가하기

# 크기 설정 (가로 14인치 × 세로 6인치)
plt.figure(figsize=(14,6))  
# 제목
plt.title("Daily Global Streams of Popular Songs in 2017-2018")  
sns.lineplot(data=spotify_data)
plt.show()
```

## 일부 열만 선택해 그리기

```python
# 컬럼 이름 확인

list(spotify_data.columns)
```

```python
# 'Shape of You'와 'Despacito' 두 곡만 그리기

plt.figure(figsize=(14,6))
plt.title("Daily Global Streams of Popular Songs in 2017-2018")

# 현재 Axes 객체 가져오기
ax = plt.gca()

# spotify_data['Shape of You'] : 특정 column 하나만 선택
# label="Shape of You" : legend에 표시될 이름 지정
sns.lineplot(data=spotify_data['Shape of You'], label="Shape of You", ax=ax)
sns.lineplot(data=spotify_data['Despacito'], label="Despacito", ax=ax)

# X축 이름 지정
plt.xlabel("Date") 

plt.show()
```
